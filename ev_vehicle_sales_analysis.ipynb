## Analysis

```python
#Q-1)List the top 3 and bottom 3 makers for the fiscal years 2023 and 2024 in terms of the number of 2-wheelers sold.
df_maker_date=pd.merge(df_ev_sales_by_makers,df_date,on='date')
```

```python
df_maker_date.head(3)
```

```python

df_maker_date['date']=pd.to_datetime(df_maker_date['date'],format="%d %B %Y")
```

```python
df_maker_date_line=df_maker_date.groupby(['date','vehicle_category'])['electric_vehicles_sold'].sum().reset_index()
plt.figure(figsize=(10,4))
sns.lineplot(x='date',y='electric_vehicles_sold',data=df_maker_date_line,hue='vehicle_category',palette='terrain',markers=["*",">"])
plt.xticks(rotation=45) 
plt.show()
```

```python
df_maker_date.info()
```

```python
df_1=df_maker_date[(df_maker_date.vehicle_category=='2-Wheelers')& (df_maker_date.fiscal_year==2023)|(df_maker_date.vehicle_category=='2-Wheelers')&(df_maker_date.fiscal_year==2024)].groupby(['fiscal_year','maker'],as_index=False).sum('electric_vehicles_sold')
df_1['sale_ranking']=df_1.groupby('fiscal_year')['electric_vehicles_sold'].rank(ascending=False).astype(int)
```

```python
def top_bottom_3(gp):
    sort_group=gp.sort_values(by='electric_vehicles_sold',ascending=False)
    return pd.concat([sort_group.head(3),sort_group.tail(3)])
    
result=df_1.groupby('fiscal_year',as_index=False).apply(top_bottom_3)
```

```python
result.set_index('fiscal_year',inplace=True)
```

```python
result
```

```python
def format_amount(amount):
    return f'{amount / 1_000:.1f}K'

plt.figure(figsize=(15,4))
plt.title("Top/Bottom maker in 2023 and 2024")
ax=sns.barplot(x='maker',hue='fiscal_year',y='electric_vehicles_sold',data=result,palette='muted')
plt.xticks(rotation=45) 
for x in ax.containers:
    ax.bar_label(x,labels=[format_amount(value) for value in x.datavalues])

plt.show()
```

**Q-3) List the states with negative penetration (decline) in EV sales from 2022 to 2024?**

```python
df_3=df_state_date[df_state_date.fiscal_year==2022]
df_3a=df_3.groupby('state')[['electric_vehicles_sold','total_vehicles_sold']].sum()
df_3a.rename(columns={'electric_vehicles_sold':'electric_vehicles_sold_2022','total_vehicles_sold':'total_vehicles_sold_2022'},inplace=True)
df_3a.reset_index(inplace=True)
df_3a['ev_penetration_rate_2022']=round((df_3a['electric_vehicles_sold_2022']/df_3a['total_vehicles_sold_2022'])*100,2)
df_3ab=df_3a[['state','ev_penetration_rate_2022']]
df_3ab.head(3)
```

```python
df_4=df_state_date[df_state_date.fiscal_year==2024]
df_4a=df_4.groupby('state')[['electric_vehicles_sold','total_vehicles_sold']].sum()
df_4a.reset_index(inplace=True)

df_4a.rename(columns={'electric_vehicles_sold':'electric_vehicles_sold_2024','total_vehicles_sold':'total_vehicles_sold_2024'},inplace=True)
df_4a['ev_penetration_rate_2024']=round((df_4a['electric_vehicles_sold_2024']/df_4a['total_vehicles_sold_2024'])*100,2)
df_4ab=df_4a[['state','ev_penetration_rate_2024']]
df_4ab.head(3)
```

```python

df_concat_3ab_4ab=pd.merge(df_3ab,df_4ab,on='state')
df_concat_3ab_4ab['change in penetration %']=df_concat_3ab_4ab['ev_penetration_rate_2024']-df_concat_3ab_4ab['ev_penetration_rate_2022']
df_concat_3ab_4ab
```

```python

sns.displot(df_concat_3ab_4ab['change in penetration %'],bins=[0,1,2,3,4,5,6,7,8,8,10],kde=True,rug=True,color='g',height=4,aspect=1.5)
plt.show()
```

**Q-8)What are the peak and low season months for EV sales based on the data from 2022 to 2024?**

```python
df_ev_sale_month_wise=df_state_date.groupby('date')['electric_vehicles_sold'].sum().reset_index()
df_ev_sale_month_wise_fiscal=pd.merge(df_ev_sale_month_wise,df_date,on='date')
df_ev_sale_month_wise_fiscal_year=df_ev_sale_month_wise_fiscal[['date','fiscal_year','electric_vehicles_sold']]
df_ev_sale_month_wise_fiscal_year.head(3)
```

```python
df_max=df_ev_sale_month_wise_fiscal_year[df_ev_sale_month_wise_fiscal_year.fiscal_year==2022].reset_index()
df_max_min=df_max[(df_max['electric_vehicles_sold']==df_max['electric_vehicles_sold'].max())|(df_max['electric_vehicles_sold']==df_max['electric_vehicles_sold'].min())]
df_max_min=df_max_min[['date','fiscal_year','electric_vehicles_sold']]
df_max_min.rename(columns={'electric_vehicles_sold':'max/min_ev_sold_2022'})
```

```python
df_max=df_ev_sale_month_wise_fiscal_year[df_ev_sale_month_wise_fiscal_year.fiscal_year==2023].reset_index()
df_max_min=df_max[(df_max['electric_vehicles_sold']==df_max['electric_vehicles_sold'].max())|(df_max['electric_vehicles_sold']==df_max['electric_vehicles_sold'].min())]
df_max_min=df_max_min[['date','fiscal_year','electric_vehicles_sold']]
df_max_min.rename(columns={'electric_vehicles_sold':'max/min_ev_sold_2023'})
```

```python
df_max=df_ev_sale_month_wise_fiscal_year[df_ev_sale_month_wise_fiscal_year.fiscal_year==2024].reset_index()
df_max_min=df_max[(df_max['electric_vehicles_sold']==df_max['electric_vehicles_sold'].max())|(df_max['electric_vehicles_sold']==df_max['electric_vehicles_sold'].min())]
df_max_min=df_max_min[['date','fiscal_year','electric_vehicles_sold']]
df_max_min.rename(columns={'electric_vehicles_sold':'max/min_ev_sold_2024'})
```
